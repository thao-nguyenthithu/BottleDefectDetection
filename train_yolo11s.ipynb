```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11s.pt") 

results = model.train(
    data="v1bottledefectdetection/data.yaml",  
    epochs=100,
    imgsz=640,  
    name="yolo11s_run_100_epochs", 
    device=0,           
    amp=False,
    workers=0,
    augment=True,       
    mosaic=1.0,         
    copy_paste=0.5
)
```
